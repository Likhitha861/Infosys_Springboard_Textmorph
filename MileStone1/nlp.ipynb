# Text Preprocessing
Text preprocessing is the process of cleaning and transforming raw text into a structured format suitable for analysis or machine learning. It typically includes tasks such as removing punctuation, converting text to lowercase, eliminating stopwords, and tokenization.

```
import re
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')

def comprehensive_preprocessing(text):
    """Complete preprocessing pipeline - converts raw text into clean, analyzable tokens"""
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)

    # Tokenize
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(nltk.corpus.stopwords.words('english'))
    filtered_tokens = [w for w in tokens if w not in stop_words and len(w) > 1]

    return filtered_tokens
```

```
# Function to read text from a file
def read_text(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

# Example with a single file
file_path = "input1.txt"   # replace with your file name
text_from_file = read_text(file_path)

# Apply preprocessing
processed_tokens = comprehensive_preprocessing(text_from_file)

print("Text Preprocessing:")
print("Original text:", text_from_file[:100] + "...")
print("Preprocessed tokens:", processed_tokens[:15])
print(f"Tokens after preprocessing: {len(processed_tokens)}")

```

#Stemmimg & Lemmatization
Stemming is the process of reducing words to their root or base form by chopping off prefixes or suffixes, often without considering proper grammar. Lemmatization, on the other hand, reduces words to their dictionary or canonical form (lemma) while considering context and part of speech, producing more linguistically accurate results.

```
from nltk.stem import PorterStemmer, WordNetLemmatizer
import pandas as pd

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to read text from a file
def read_text(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

# List of your text files
file_paths = ["input1.txt", "input2.txt"]

for file_path in file_paths:
    # Read and preprocess text
    text = read_text(file_path)
    processed_tokens = comprehensive_preprocessing(text)

    # Apply stemming and lemmatization
    stemmed = [stemmer.stem(w) for w in processed_tokens]
    lemmatized = [lemmatizer.lemmatize(w) for w in processed_tokens]

    # Create comparison DataFrame
    comparison_df = pd.DataFrame({
        'Original': processed_tokens,
        'Stemmed': stemmed,
        'Lemmatized': lemmatized
    })

    # Display results
    print(f"\n--- Stemming vs Lemmatization for {file_path} ---")
    print(comparison_df.head(15))  # show first 15 tokens
```

#Part-of-Speech Tagging & Named Entity Recognition
Part-of-Speech (POS) tagging is the process of labeling each word in a text with its grammatical role, such as noun, verb, adjective, or adverb. Named Entity Recognition (NER) identifies and classifies key entities in text, such as people, organizations, locations, dates, and other proper nouns.

```
# Function to read text from a file
def read_text(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

# List of your text files
file_paths = ["input1.txt", "input2.txt"]

for file_path in file_paths:
    # Read and preprocess text
    text = read_text(file_path)
    processed_tokens = comprehensive_preprocessing(text)

    # Part-of-Speech Tagging (first 15 tokens)
    pos_tags = nltk.pos_tag(processed_tokens[:15])
    print(f"\n--- Part-of-Speech Tags for {file_path} ---")
    for word, tag in pos_tags:
        print(f"{word}: {tag}")

    # Named Entity Recognition using spaCy
    doc = nlp(text)
    print(f"\nNamed Entities in {file_path}:")
    if doc.ents:
        for ent in doc.ents:
            print(f"{ent.text}: {ent.label_} ({spacy.explain(ent.label_)})")
    else:
        print("No named entities found in this text")

    # Dependency Parsing (first 10 tokens)
    print(f"\nDependency Relations (first 10 tokens) in {file_path}:")
    for token in doc[:10]:
        print(f"{token.text} -> {token.dep_} -> {token.head.text}")
```

#Frequency Analysis & Visualization
Frequency analysis involves counting how often each word or token appears in a text, helping identify the most common terms or patterns. Visualization, such as bar charts or word clouds, provides a graphical representation of these frequencies, making it easier to interpret and communicate insights from the text.

```
from collections import Counter
import matplotlib.pyplot as plt
from nltk import FreqDist

# List of your text files
file_paths = ["input1.txt", "input2.txt"]

for file_path in file_paths:
    # Read and preprocess text
    text = read_text(file_path)
    processed_tokens = comprehensive_preprocessing(text)

    # Word frequency analysis
    word_freq = Counter(processed_tokens)
    most_common = word_freq.most_common(10)
    print(f"\n--- Top 10 Word Frequencies for {file_path} ---")
    for word, count in most_common:
        print(f"{word}: {count}")

    # Bar chart of top 10 words
    if most_common:
        words, counts = zip(*most_common)
        plt.figure(figsize=(12, 6))
        plt.bar(words, counts, color='skyblue', alpha=0.8)
        plt.title(f'Top 10 Word Frequencies - {file_path}', fontsize=16)
        plt.xlabel('Words', fontsize=12)
        plt.ylabel('Frequency', fontsize=12)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    # Frequency distribution plot for all words
    fdist = FreqDist(processed_tokens)
    plt.figure(figsize=(12, 6))
    fdist.plot(20, title=f"Word Frequency Distribution - {file_path}")
    plt.show()
```

#Text Readability Analysis
Readability metrics measure how easy text is to read and understand, useful for content optimization.

```
import textstat

# List of your text files
file_paths = ["input1.txt", "input2.txt"]

for file_path in file_paths:
    text = read_text(file_path)  # Read the text from the file

    print(f"\n--- Readability Metrics for {file_path} ---")
    try:
        flesch = textstat.flesch_reading_ease(text)
        fk_grade = textstat.flesch_kincaid_grade(text)
        gunning = textstat.gunning_fog(text)
        ari = textstat.automated_readability_index(text)
        coleman = textstat.coleman_liau_index(text)

        print(f"Flesch Reading Ease: {flesch:.2f} (Higher = easier)")
        print(f"Flesch-Kincaid Grade: {fk_grade:.2f} (US grade level)")
        print(f"Gunning Fog Index: {gunning:.2f} (Years of education needed)")
        print(f"Automated Readability Index: {ari:.2f}")
        print(f"Coleman-Liau Index: {coleman:.2f}")

        # Interpretation based on Flesch Reading Ease
        if flesch >= 90:
            difficulty = "Very Easy"
        elif flesch >= 80:
            difficulty = "Easy"
        elif flesch >= 70:
            difficulty = "Fairly Easy"
        elif flesch >= 60:
            difficulty = "Standard"
        elif flesch >= 50:
            difficulty = "Fairly Difficult"
        elif flesch >= 30:
            difficulty = "Difficult"
        else:
            difficulty = "Very Difficult"

        print(f"Interpretation: {difficulty}")

    except Exception as e:
        print(f"Readability analysis error: {e}")
```

#Reading and Processing Online Text Files
Loading text from URLs enables analysis of web-hosted documents like books, articles, and datasets.

```
# Fetch text file from Project Gutenberg - classic literature repository
print("Downloading online text file...")
txt_url = "https://www.gutenberg.org/files/84/84-0.txt"  # Frankenstein

try:
    response = requests.get(txt_url, timeout=30)
    response.raise_for_status()  # Raise an exception for bad status codes
    book_text = response.text

    print(f"Text file loaded successfully!")
    print(f"File length: {len(book_text):,} characters")
    print("Preview:", book_text[:300] + "...")

except requests.exceptions.RequestException as e:
    print(f"Error downloading file: {e}")
    # Fallback text for demonstration
    book_text = """This is a sample text for demonstration purposes when the online file cannot be downloaded.
    It contains multiple sentences and various words that can be analyzed using NLP techniques.
    Natural language processing is fascinating and useful for many applications."""
    print("Using fallback text for demonstration")

# Clean and process the book text - removes Project Gutenberg headers/footers
def clean_gutenberg_text(text):
    """Clean Project Gutenberg text by removing metadata"""
    # Look for content markers
    start_markers = ["*** START OF", "***START OF", "CHAPTER", "Chapter"]
    end_markers = ["*** END OF", "***END OF"]

    # Find start of actual content
    start_idx = 0
    for marker in start_markers:
        idx = text.find(marker)
        if idx != -1:
            # Find end of line after marker
            start_idx = text.find('\n', idx)
            break

    # Find end of content
    end_idx = len(text)
    for marker in end_markers:
        idx = text.find(marker)
        if idx != -1:
            end_idx = idx
            break

    # Extract main content
    if start_idx > 0:
        text = text[start_idx:end_idx]

    return text.strip()

print("\nProcessing text...")
cleaned_book = clean_gutenberg_text(book_text)
book_tokens = comprehensive_preprocessing(cleaned_book)

print(f"Book processed: {len(book_tokens):,} tokens")
```

```
# Comprehensive analysis of Amazon Reviews dataset
if csv_tokens and main_text_col in df.columns:
    print(f"\n{'='*70}")
    print(f"COMPREHENSIVE AMAZON REVIEWS ANALYSIS")
    print(f"{'='*70}")

    # === DATASET OVERVIEW ===
    print(f"\nDataset Overview:")
    print(f"- Total reviews: {len(df):,}")
    print(f"- Total tokens processed: {len(csv_tokens):,}")
    print(f"- Unique words: {len(set(csv_tokens)):,}")
    print(f"- Vocabulary richness: {len(set(csv_tokens))/len(csv_tokens):.3f}")

    # === WORD FREQUENCY ANALYSIS ===
    print(f"\nTop Words in Amazon Reviews:")
    csv_freq = Counter(csv_tokens)
    top_words = csv_freq.most_common(25)

    for i, (word, count) in enumerate(top_words[:15], 1):
        percentage = (count / len(csv_tokens)) * 100
        print(f"{i:2d}. {word:<15} {count:>5} times ({percentage:>5.1f}%)")

    # === RATING ANALYSIS ===
    print(f"\nRating Analysis:")
    if rating_column in df.columns:
        rating_stats = df[rating_column].describe()
        rating_dist = df[rating_column].value_counts().sort_index()

        print(f"Rating Statistics:")
        print(f"- Average rating: {rating_stats['mean']:.2f} stars")
        print(f"- Most common rating: {rating_dist.idxmax()} stars")
        print(f"- Rating standard deviation: {rating_stats['std']:.2f}")

        print(f"\nRating Distribution:")
        for rating, count in rating_dist.items():
            percentage = (count / len(df)) * 100
            stars = "★" * int(rating) + "☆" * (5 - int(rating))
            print(f"  {rating} {stars}: {count:4d} reviews ({percentage:5.1f}%)")

    # === SENTIMENT ANALYSIS ===
    print(f"\nSentiment Analysis:")

    # Create sentiment labels based on ratings
    if rating_column in df.columns:
        def rating_to_sentiment(rating):
            if rating >= 4:
                return 'positive'
            elif rating <= 2:
                return 'negative'
            else:
                return 'neutral'

        df['sentiment'] = df[rating_column].apply(rating_to_sentiment)
        sentiment_dist = df['sentiment'].value_counts()

        print("Sentiment Distribution (based on ratings):")
        for sentiment, count in sentiment_dist.items():
            percentage = (count / len(df)) * 100
            print(f"  {sentiment.capitalize()}: {count:4d} reviews ({percentage:5.1f}%)")

    # === VISUALIZATIONS ===
    print(f"\nCreating visualizations...")

    # 1. Rating Distribution
    if rating_column in df.columns:
        plt.figure(figsize=(12, 6))
        bars = plt.bar(rating_dist.index, rating_dist.values,
                      color='gold', alpha=0.8, edgecolor='orange', linewidth=2)
        plt.title('Amazon Reviews - Rating Distribution', fontsize=16, pad=20)
        plt.xlabel('Rating (Stars)', fontsize=12)
        plt.ylabel('Number of Reviews', fontsize=12)
        plt.xticks([1, 2, 3, 4, 5])
        plt.grid(True, alpha=0.3)

        # Add value labels on bars
        for bar, count in zip(bars, rating_dist.values):
            plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + max(rating_dist.values)*0.01,
                    f'{count}', ha='center', va='bottom', fontsize=11, fontweight='bold')

        plt.tight_layout()
        plt.show()

    # 2. Sentiment Distribution
    if 'sentiment' in df.columns:
        plt.figure(figsize=(10, 6))
        colors = {'positive': 'green', 'negative': 'red', 'neutral': 'gray'}
        bar_colors = [colors.get(sent, 'blue') for sent in sentiment_dist.index]

        bars = plt.bar(sentiment_dist.index, sentiment_dist.values,
                      color=bar_colors, alpha=0.7)
        plt.title('Amazon Reviews - Sentiment Distribution', fontsize=16)
        plt.xlabel('Sentiment', fontsize=12)
        plt.ylabel('Number of Reviews', fontsize=12)

        # Add value labels
        for bar, count in zip(bars, sentiment_dist.values):
            plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + max(sentiment_dist.values)*0.01,
                    f'{count}', ha='center', va='bottom', fontsize=12, fontweight='bold')

        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    # 3. Top Words Bar Chart
    if len(top_words) > 0:
        words, counts = zip(*top_words[:15])
        plt.figure(figsize=(16, 8))
        bars = plt.bar(words, counts, color='lightcoral', alpha=0.8)
        plt.title('Top 15 Words in Amazon Reviews', fontsize=18, pad=20)
        plt.xlabel('Words', fontsize=14)
        plt.ylabel('Frequency', fontsize=14)
        plt.xticks(rotation=45, ha='right', fontsize=12)

        # Add value labels
        for bar, count in zip(bars, counts):
            plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + max(counts)*0.01,
                    f'{count}', ha='center', va='bottom', fontsize=11, fontweight='bold')

        plt.tight_layout()
        plt.show()

    # 4. Review Length Distribution
    if review_lengths:
        plt.figure(figsize=(14, 6))
        plt.hist(review_lengths, bins=30, color='skyblue', alpha=0.7, edgecolor='black')
        plt.title('Distribution of Amazon Review Lengths', fontsize=16)
        plt.xlabel('Number of Tokens per Review', fontsize=12)
        plt.ylabel('Number of Reviews', fontsize=12)

        avg_length = sum(review_lengths) / len(review_lengths)
        plt.axvline(avg_length, color='red', linestyle='--', linewidth=2,
                    label=f'Average: {avg_length:.1f} tokens')
        plt.legend(fontsize=12)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    # 5. Word Cloud
    print(f"Creating Amazon Reviews word cloud...")
    if len(csv_tokens) > 50:
        amazon_wordcloud = WordCloud(
            width=1400,
            height=700,
            background_color='white',
            colormap='viridis',
            max_words=120,
            stopwords=STOPWORDS.union({'product', 'amazon', 'buy', 'purchase', 'item'}),
            collocations=False,
            prefer_horizontal=0.6
        ).generate(" ".join(csv_tokens))

        plt.figure(figsize=(18, 9))
        plt.imshow(amazon_wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud - Amazon Product Reviews', fontsize=20, pad=20)
        plt.show()

    # === ADVANCED SENTIMENT ANALYSIS ===
    try:
        from nltk.sentiment import SentimentIntensityAnalyzer
        sia = SentimentIntensityAnalyzer()

        print(f"\nAdvanced Sentiment Analysis (VADER):")
        print("Analyzing sample reviews with VADER sentiment analyzer...")

        # Analyze sentiment for sample reviews
        sample_reviews = df[main_text_col].fillna('').head(5).tolist()
        sample_ratings = df[rating_column].head(5).tolist() if rating_column in df.columns else [None]*5

        for i, (review, rating) in enumerate(zip(sample_reviews, sample_ratings), 1):
            if len(str(review)) > 20:
                scores = sia.polarity_scores(str(review))
                compound = scores['compound']

                if compound >= 0.05:
                    vader_sentiment = "Positive"
                elif compound <= -0.05:
                    vader_sentiment = "Negative"
                else:
                    vader_sentiment = "Neutral"

                print(f"\n{i}. Rating: {rating}⭐ | VADER: {vader_sentiment} ({compound:.3f})")
                print(f"   '{str(review)[:120]}{'...' if len(str(review)) > 120 else ''}'")

    except Exception as e:
        print(f"VADER sentiment analysis not available: {e}")

    # === N-GRAMS ANALYSIS ===
    print(f"\nN-grams Analysis:")
    from nltk import bigrams, trigrams

    # Bigrams analysis
    bigrams_list = list(bigrams(csv_tokens))
    if bigrams_list:
        bigram_freq = Counter(bigrams_list)
        print("Top 8 bigrams (word pairs) in Amazon reviews:")
        for i, (bigram, count) in enumerate(bigram_freq.most_common(8), 1):
            print(f"  {i}. '{' '.join(bigram)}' - {count} occurrences")

    # Trigrams analysis
    trigrams_list = list(trigrams(csv_tokens))
    if trigrams_list:
        trigram_freq = Counter(trigrams_list)
        print(f"\nTop 5 trigrams (word triplets) in Amazon reviews:")
        for i, (trigram, count) in enumerate(trigram_freq.most_common(5), 1):
            print(f"  {i}. '{' '.join(trigram)}' - {count} occurrences")

    # === PRODUCT ANALYSIS ===
    if 'asin' in df.columns:
        print(f"\nProduct Analysis:")
        product_counts = df['asin'].value_counts()
        print(f"Number of unique products: {len(product_counts)}")
        print(f"Most reviewed products:")

        for i, (asin, count) in enumerate(product_counts.head(5).items(), 1):
            avg_rating = df[df['asin'] == asin][rating_column].mean() if rating_column in df.columns else 'N/A'
            print(f"  {i}. Product {asin}: {count} reviews (avg rating: {avg_rating:.1f}⭐)")

    # === SAMPLE REVIEWS SHOWCASE ===
    print(f"\nSample Amazon Reviews:")
    print("Representative reviews from the dataset:")

    sample_indices = [0, len(df)//4, len(df)//2]  # Show 3 samples

    for i, idx in enumerate(sample_indices, 1):
        if idx < len(df):
            review = str(df[main_text_col].iloc[idx])
            rating = df[rating_column].iloc[idx] if rating_column in df.columns else 'N/A'
            summary = str(df['summary'].iloc[idx]) if 'summary' in df.columns else 'N/A'
            reviewer = str(df[reviewer_column].iloc[idx]) if reviewer_column in df.columns else 'Anonymous'

            print(f"\n--- Review {i} ---")
            print(f"Reviewer: {reviewer}")
            print(f"Rating: {rating}⭐")
            print(f"Summary: \"{summary}\"")
            print(f"Review: \"{review[:250]}{'...' if len(review) > 250 else ''}\"")

else:
    print("No suitable Amazon review data found for analysis")
```

```
# Compare different texts - quantitative analysis of text characteristics
def text_summary(tokens, name):
    """Generate comprehensive summary statistics for a text"""
    if not tokens:
        return {
            'name': name,
            'total_words': 0,
            'unique_words': 0,
            'vocabulary_richness': 0,
            'avg_word_length': 0,
            'most_common_word': ('N/A', 0)
        }

    freq_dist = Counter(tokens)
    return {
        'name': name,
        'total_words': len(tokens),
        'unique_words': len(set(tokens)),
        'vocabulary_richness': round(len(set(tokens)) / len(tokens), 3),
        'avg_word_length': round(sum(len(word) for word in tokens) / len(tokens), 2),
        'most_common_word': freq_dist.most_common(1)[0] if freq_dist else ('N/A', 0)
    }

# Compare all available texts
print("Comparative Text Analysis:")
texts_comparison = [
    text_summary(processed_tokens, "Sample Text"),
    text_summary(csv_tokens[:5000] if len(csv_tokens) > 5000 else csv_tokens, "Amazon Reviews")
]

comparison_df = pd.DataFrame(texts_comparison)
print(comparison_df.to_string(index=False))

# Visualize comparison
metrics = ['total_words', 'unique_words', 'vocabulary_richness', 'avg_word_length']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

for i, metric in enumerate(metrics):
    names = comparison_df['name'].tolist()
    values = comparison_df[metric].tolist()

    axes[i].bar(names, values, color=['skyblue', 'lightcoral'], alpha=0.8)
    axes[i].set_title(f'{metric.replace("_", " ").title()}', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
